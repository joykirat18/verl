```python
import pandas as pd
```

```python
## open parquet file
file_path = "/nas-ssd2/joykirat/code/latent-reasoning-project/verl/scripts/data/latr-data/countdown-base/test.parquet"  # replace with the path to your parquet file
df = pd.read_parquet(file_path)
print(df.head())
```

```python
len(df)
```

```python
df_small = df.head(250)
```

```python
# Iterate through the DataFrame, edit the 'prompt', filter/cut down rows, and save a new parquet

# Example: edit 'prompt' column
from tqdm import tqdm
for idx in tqdm(range(len(df_small))):
    nums = df_small.at[idx, "nums"]
    target = df_small.at[idx, "target"]

    prompt = f"Using the numbers {nums}, create an equation that equals {target}. You can use basic arithmetic operations (+, -, *, /) and each number can only be used once. Return the final answer in <answer> </answer> tags, for example <answer> (1 + 2) / 3 </answer>. Let's think step by step and output the final answer within <answer> </answer> tags."

    prompt = [{"role": "user", "content": prompt}]

    df_small.at[idx, "prompt"] = prompt
```

```python
output_path = file_path.replace(".parquet", "_v2.parquet")
```

```python
print(output_path)
```

```python
df_small.to_parquet(output_path)
```
