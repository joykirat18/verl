```python
import json
## read jsonl file
with open('overthink_bench.jsonl', 'r') as f:
    data = [json.loads(line) for line in f]
```

```python
filtered_data = []
for d in data:
    if d['mode'] == 'numeric' or d['mode'] == 'mcq':
        filtered_data.append(d)
```

```python
import json
with open('filtered_overthink_bench.json', 'w') as f:
    json.dump(filtered_data, f, indent=4)
```

```python
import uuid

# Add a 'uuid' column to the DataFrame, assigning a new UUID to each row
data['uuid'] = [str(uuid.uuid4()) for _ in range(len(data))]


```

```python
# Convert the DataFrame 'data' to JSON format
data_json = data.to_json(orient='records', indent=4)


```

```python
# Save the JSON data to a file
with open('underthink_bench.json', 'w') as f:
    f.write(data_json)
```
